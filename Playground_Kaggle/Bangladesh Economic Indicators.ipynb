# Bangladesh Economic Indicators (1980 - 2019)
## Data visualization

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/BD_economic_indicators.csv')
df.head()
```

```python
df.dtypes
```

```python
def convertStrWithComma2Int(x):
    if type(x) is str:
        return int(x.replace(',', ''))
    else:
        return x

def convertStrWithPercent2Float(x):
    if type(x) is str:
        return float(x[:-1])
    else:
        return x
```

```python
df['GDP per capita'] = df['GDP per capita'].map(convertStrWithComma2Int)
df['GDP growth'] = df['GDP growth'].map(convertStrWithPercent2Float)
df['Inflation rate'] = df['Inflation rate'].map(convertStrWithPercent2Float)
df['Total Investment'] = df['Total Investment'].map(convertStrWithPercent2Float)
df['Unemployed rate'] = df['Unemployed rate'].map(convertStrWithPercent2Float)
df['Government debt'] = df['Government debt'].map(convertStrWithPercent2Float)
df.columns
```

```python
gdpTransition = sns.lineplot(x='Year', y='GDP ', data=df)
gdpTransition.set_title('GDP Transition (billion USD)')
```

```python
perCapitaGDPTransition = sns.lineplot(x='Year', y='GDP per capita', data=df)
perCapitaGDPTransition.set_title('Per capita GDP Transition($)')
```

```python
gdpGrowth = sns.lineplot(x='Year', y='GDP growth', data=df)
gdpGrowth.set_title('GDP Growth(%)')
```

```python
gdpInflationRate = sns.lineplot(x='Year', y='Inflation rate', data=df)
gdpInflationRate.set_title('Inflation Rate(%)')
```

```python
unemploymentRate = sns.lineplot(x='Year', y='Unemployed rate', data=df)
unemploymentRate.set_title('Unemployment Rate(%)')
```

```python
governmentDebt = sns.lineplot(x='Year', y='Government debt', data=df)
governmentDebt.set_title('Government Debt(%)')
```

```python
totalInvestment = sns.lineplot(x='Year', y='Total Investment', data=df)
totalInvestment.set_title('Total Investment(%)')
```
