```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
```

```python
# Load an image in grayscale
img_gray = cv2.imread('test_image.jpg', cv2.IMREAD_GRAYSCALE)

# Display the image
#cv2.imshow('Image', img_gray)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

plt.imshow(img_gray)
plt.title('Grayscaled Image')
plt.show()
```
